```python
# Bienvenida e inicialización
print("--- Bienvenido a IA 'BeautyMaster' ---")
print("Tu consultor inteligente para un cambio de imagen total.")

# Captura del presupuesto inicial (Operador de asignación) [4]
presupuesto = float(input("\nPara comenzar, ¿cuánto dinero tienes para tu imagen hoy?: "))

# Inicio del ciclo infinito para el menú interactivo [1, 5]
while True:
    print("\n--- MENÚ DE SERVICIOS INTELIGENTES ---")
    print("1. Sugerencia de Corte ($30)")
    print("2. Recomendación de Tinte ($70)")
    print("3. Tratamiento Reparador ($50)")
    print("4. Consultar Saldo Disponible")
    print("5. Salir")

    opcion = input("\nSelecciona un servicio (1-5): ")

    # Opción 1: Lógica de Corte con operadores relacionales y lógicos [2, 3]
    if opcion == "1":
        if presupuesto >= 30:
            print("\nIA: Analizando estructura ósea...")
            forma_rostro = input("¿Tu rostro es 'Redondo' u 'Ovalado'?: ")

            # Decisión basada en comparación de cadenas y operador lógico 'or' [2]
            if forma_rostro == "Redondo" or forma_rostro == "redondo":
                print("Recomendación IA: Un corte 'Pixie' con volumen arriba para alargar tu rostro.")
            elif forma_rostro == "Ovalado" or forma_rostro == "ovalado":
                print("Recomendación IA: Un corte en capas largas resaltará tu simetría.")
            else:
                print("Recomendación IA: Un estilo 'Bob' clásico es tu mejor opción.")

            # Operador aritmético de resta para actualizar el presupuesto [4]
            presupuesto = presupuesto - 30
            print(f"¡Corte realizado! Saldo restante: ${presupuesto}")
        else:
            print("\nIA: Error. Fondos insuficientes para este servicio.")

    # Opción 2: Lógica de Tinte con evaluación de múltiples condiciones [6, 7]
    elif opcion == "2":
        if presupuesto >= 70:
            print("\nIA: Escaneando tono de piel...")
            tono = input("¿Tu piel es de tono 'Cálido' o 'Frío'?: ")

            if tono == "Cálido" or tono == "calido":
                print("Recomendación IA: Los tonos cobrizos o dorados iluminarán tu cara.")
            elif tono == "Frío" or tono == "frio":
                print("Recomendación IA: Los tonos cenizos o platinados te darán un look elegante.")
            else:
                print("Recomendación IA: Un castaño chocolate es ideal para ti.")

            presupuesto = presupuesto - 70
            print(f"¡Color aplicado! Saldo restante: ${presupuesto}")
        else:
            print("\nIA: Lo sentimos, necesitas más presupuesto para el tinte.")

    # Opción 3: Lógica de Tratamiento con operador relacional de comparación [2, 8]
    elif opcion == "3":
        if presupuesto >= 50:
            print("\nIA: Evaluando salud capilar...")
            danio = int(input("Del 1 al 10, ¿qué tan maltratado sientes tu cabello?: "))

            # Uso de operador mayor o igual que (>=) [2]
            if danio >= 7:
                print("Recomendación IA: Aplicaremos una máscara de keratina profunda.")
            else:
                print("Recomendación IA: Una hidratación ligera con aceites naturales será suficiente.")

            presupuesto = presupuesto - 50
            print(f"¡Tratamiento finalizado! Saldo restante: ${presupuesto}")
        else:
            print("\nIA: Presupuesto insuficiente para tratamiento.")

    # Opción 4: Consulta simple de saldo [9, 10]
    elif opcion == "4":
        print(f"\nTu saldo actual es de: ${presupuesto}")

    # Opción 5: Salida controlada del bucle [11, 12]
    elif opcion == "5":
        print("\nIA BeautyMaster: Gracias por tu visita. ¡Disfruta tu nuevo look!")
        break # Termina la ejecución del ciclo while True [13]

    # Manejo de entradas inválidas [14, 15]
    else:
        print("\nError: Por favor selecciona una opción válida del 1 al 5.")
```
